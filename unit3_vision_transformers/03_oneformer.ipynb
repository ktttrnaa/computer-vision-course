# OneFormer: Universal Image Segmentation

## Goal

The purpose of this notebook is to demonstrate the main idea behind OneFormer.

OneFormer is a universal segmentation model that can work with different segmentation tasks using a single architecture.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## Simple Scene

First, let's create a simple artificial image.

The upper part represents the sky, and the lower part represents grass.

```python
image = mpimg.imread("cat.png")

plt.figure(figsize=(6,6))
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

## Semantic Segmentation

Semantic segmentation assigns a class label to every pixel in the image.

In this example we will simulate semantic segmentation using three classes:

- Sky
- Vegetation
- Cat

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image = mpimg.imread("cat.png")

mask = np.zeros((image.shape[0], image.shape[1]))

# sky
mask[:220, :] = 1

# vegetation
mask[220:, :] = 2

# cat
mask[40:340, 220:400] = 3

plt.figure(figsize=(6,6))
plt.imshow(mask)
plt.title("Semantic Segmentation")
plt.axis("off")
plt.show()
```

## Instance Segmentation

Instance segmentation identifies individual objects separately.

Unlike semantic segmentation, different objects of the same class receive different labels.

```python
instance_mask = np.zeros((image.shape[0], image.shape[1]))

# cat instance
instance_mask[40:340, 220:400] = 1

plt.figure(figsize=(6,6))
plt.imshow(instance_mask)
plt.title("Instance Segmentation")
plt.axis("off")
plt.show()
```

## Panoptic Segmentation

Panoptic segmentation combines semantic segmentation and instance segmentation.

Background regions are assigned semantic labels, while individual objects receive separate instance labels.

This unified approach is one of the key ideas behind OneFormer.

```python
panoptic_mask = np.zeros((image.shape[0], image.shape[1]))

# sky
panoptic_mask[:220, :] = 1

# vegetation
panoptic_mask[220:, :] = 2

# cat instance
panoptic_mask[40:340, 220:400] = 3

plt.figure(figsize=(6,6))
plt.imshow(panoptic_mask)
plt.title("Panoptic Segmentation")
plt.axis("off")
plt.show()
```

## Why OneFormer?

Traditional approaches often require separate models for:

- semantic segmentation;
- instance segmentation;
- panoptic segmentation.

OneFormer introduces a unified architecture that can perform all these tasks using a single model.

The desired task is selected dynamically through task conditioning.
