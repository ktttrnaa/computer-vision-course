# MobileViT v2

## Goal

The purpose of this notebook is to demonstrate the main idea behind MobileViT v2.

MobileViT v2 combines the strengths of Convolutional Neural Networks and Vision Transformers.

It is designed for mobile and resource-constrained devices.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## CNN, ViT and MobileViT

CNNs are efficient and good at extracting local features.

Vision Transformers can model global relationships, but they are usually computationally expensive.

MobileViT combines local CNN-style processing with global Transformer-style processing.

```python
models = {
    "MobileNet": "CNN",
    "ViT": "Transformer",
    "MobileViT": "CNN + Transformer"
}

for name, kind in models.items():
    print(f"{name}: {kind}")
```

## Local Features and Global Context

CNNs are good at extracting local features from small image regions.

Transformers are useful for modeling global relationships across the whole image.

MobileViT combines both ideas.

```python
image = mpimg.imread("cat.png")

plt.figure(figsize=(6, 4))
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
local_region = image[80:260, 220:390]

plt.figure(figsize=(4, 4))
plt.imshow(local_region)
plt.title("Local Region")
plt.axis("off")
plt.show()
```

## Global Context

While CNNs focus on local image regions, Transformers can model relationships across the entire image.

This allows the model to understand both object details and the surrounding scene.

```python
plt.figure(figsize=(8, 3))

plt.subplot(1, 2, 1)
plt.imshow(local_region)
plt.title("Local Features")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(image)
plt.title("Global Context")
plt.axis("off")

plt.tight_layout()
plt.show()
```

## MobileViT Block

The MobileViT block combines local feature extraction from CNNs with global feature modeling from Transformers.

The workflow is:

Image → Convolution → Patches → Transformer → Fusion

This allows MobileViT to capture both local and global information efficiently.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 2))

plt.text(0.05, 0.5, "Image", fontsize=12)
plt.text(0.25, 0.5, "Conv", fontsize=12)
plt.text(0.45, 0.5, "Patches", fontsize=12)
plt.text(0.65, 0.5, "Transformer", fontsize=12)
plt.text(0.88, 0.5, "Fusion", fontsize=12)

plt.arrow(0.12, 0.5, 0.08, 0,
          head_width=0.02,
          head_length=0.02,
          length_includes_head=True)

plt.arrow(0.32, 0.5, 0.08, 0,
          head_width=0.02,
          head_length=0.02,
          length_includes_head=True)

plt.arrow(0.52, 0.5, 0.08, 0,
          head_width=0.02,
          head_length=0.02,
          length_includes_head=True)

plt.arrow(0.78, 0.5, 0.06, 0,
          head_width=0.02,
          head_length=0.02,
          length_includes_head=True)

plt.axis("off")
plt.title("MobileViT Block")
plt.show()
```

## Model Size Comparison

One of the main goals of MobileViT is to provide Transformer-level performance while remaining efficient enough for mobile devices.

Compared to larger architectures, MobileViT uses significantly fewer parameters.

```python
import matplotlib.pyplot as plt

models = [
    "MobileNetV2",
    "ResNet18",
    "ViT"
]

parameters = [
    3.5,
    11.7,
    86.4
]

plt.figure(figsize=(7,4))
plt.bar(models, parameters)

plt.ylabel("Millions of Parameters")
plt.title("Efficiency of Mobile Vision Models")

plt.show()
```
